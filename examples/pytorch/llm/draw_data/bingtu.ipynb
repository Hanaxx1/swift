```python
#欢迎帅哥
#开发时间: 2024-02-01 13:08

from collections import Counter



origin_counter = Counter()
destination_counter = Counter()

country_name_tran_dict = {'United States':'the United States','Korea':'South Korea','the Congo':'Congo','DR Congo':'The Democratic Republic of Congo',
                          'Congo Brazzaville':'The Democratic Republic of Congo','St. Lucia':'Saint Lucia'}

def country_name_tran(country_name):
    if country_name in country_name_tran_dict:
        return country_name_tran_dict[country_name]
    else:
        return country_name
    


def get_characters_before_third_comma(input_string):
    count_comma = 0
    index_third_comma = None

    for i, char in enumerate(input_string):
        if char == ',':
            count_comma += 1
            if count_comma == 3:
                index_third_comma = i
                break

    if index_third_comma is not None:
        return input_string[:index_third_comma]
    else:
        return "Third comma not found"

import json
import matplotlib.pyplot as plt
import linecache

path_file = '/home/qiuyang/workplace/swift/examples/pytorch/llm/my_data/data/c4_data_duiqi/c4_data.json'

count = 0

with open(path_file, 'r') as input_file:
    # 逐行读取文件内容
    for line in input_file:
        # 解析JSON数据
        data = json.loads(line)

        origin = country_name_tran(data['origin'])
        destination = country_name_tran(data['destination'])

        # 统计出现次数
        origin_counter[origin] += 1
        destination_counter[destination] += 1

# 按值从大到小排序
sorted_origin = dict(sorted(origin_counter.items(), key=lambda item: item[1], reverse=True))
sorted_destination = dict(sorted(destination_counter.items(), key=lambda item: item[1], reverse=True))

print("Sorted Origin:")
for key, value in sorted_origin.items():
    print(f"{key}: {value}")

print("\nSorted Destination:")
for key, value in sorted_destination.items():
    print(f"{key}: {value}")
```

```python
total_sum = sum(sorted_destination.values())

plot_bin_data_origin = {'Others':0}

for key, value in sorted_origin.items():
    if value > 410:
        plot_bin_data_origin[key] = round(value/total_sum,3)
    else:
        plot_bin_data_origin['Others'] += value

plot_bin_data_origin['Others'] = round(plot_bin_data_origin['Others']/total_sum,2)
print(len(plot_bin_data_origin))

print("Total sum of values:", total_sum)
print("Updated plot_bin_data:", plot_bin_data_origin)


plot_bin_data_origin_to_destination = {'Others':0}
for key, value in sorted_destination.items():
    if key in plot_bin_data_origin:
        plot_bin_data_origin_to_destination[key] = round(value/total_sum,3)
    else:
        plot_bin_data_origin_to_destination['Others'] += value

plot_bin_data_origin_to_destination['Others'] = round(plot_bin_data_origin_to_destination['Others']/total_sum,2)
# print(plot_bin_data_origin_to_destination)

plot_bin_data_origin_to_destination_sort = {}


for key,value in plot_bin_data_origin_to_destination.items():
    plot_bin_data_origin_to_destination_sort[key] = value

print(plot_bin_data_origin_to_destination_sort)

plot_bin_data_destination = {'Others':0}

for key, value in sorted_destination.items():
    if value > 256:
        plot_bin_data_destination[key] = round(value/total_sum,3)
    else:
        plot_bin_data_destination['Others'] += value


plot_bin_data_destination['Others'] = round(plot_bin_data_destination['Others']/total_sum,2)


print(len(plot_bin_data_destination))
print("Total sum of values:", total_sum)
print("Updated plot_bin_data:", plot_bin_data_destination)
```
